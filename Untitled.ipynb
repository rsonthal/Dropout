```julia
function minibatch(x,y,batchs)
    numBatch = Int((size(x)[2])/batchs)
    b = Array{Any,1}(numBatch)
    for i = 1:numBatch
        s = (i-1)*batchs + 1
        e = s + batchs -1
        b[i] = (x[:,s:e],y[:,s:e])
    end
    return b
end
```

```julia
function zeroone(data,model)
    ncorr = ninst = 0
    for (x,y) in data
        ypred = model(x)
        ncorr += sum(y .* (ypred .== maximum(ypred,1)))
        ninst += size(x,2)
    end
    return 1 - ncorr/ninst
end
```

# Measuring Loss

```julia
# Calculate cross entropy loss of a model with weights w for one minibatch (x,p)
# Use non-zero l1 or l2 for regularization
function softloss(x,p,m;l1=0,l2=0)  
    y = m(x).data
    y = y .- maximum(y,1) # for numerical stability
    expy = exp.(y)
    logphat = y .- log.(sum(expy,1))
    J = -sum(p .* logphat) / size(x,2)  # dividing by number of instances for per-instance average
    if l1 != 0; J += l1 * sum(vecnorm, params(m)); end
    if l2 != 0; J += l2 * sum(vecnorm, params(m)); end
    return J
end

# Calculate cross entropy loss of a model with weights w for one minibatch (x,p)
# Use non-zero l1 or l2 for regularization
function softloss2(x,p,m;l1=0,l2=0)  
    y = m(x)
    y = y .- maximum(y,1) # for numerical stability
    expy = exp.(y)
    logphat = y .- log.(sum(expy,1))
    J = -sum(p .* logphat) / size(x,2)  # dividing by number of instances for per-instance average
    if l1 != 0; J += l1 * sum(vecnorm, params(m)); end
    if l2 != 0; J += l2 * sum(vecnorm, params(m)); end
    return J
end
```

```julia
function avgloss(data,m) # average loss for the whole dataset
    sum = cnt = 0
    for (x,y) in data
        ypred = m(x).data
        sum += vecnorm(ypred-y,2)/100
        cnt += 1
    end
    return sum/cnt
end
```

# Training (SGD)

```julia
m = Chain(Dense(784,10))
loss(r,s) = vecnorm(m(r)-s,2)
```

```julia
using Flux.Tracker
using Flux: @epochs
```

# Training the Linear Model

```julia
trnloss1 = zeros(100)
tstloss1 = zeros(100)
trnerr1 = zeros(100)
tsterr1 = zeros(100)
for i = 1:100
    Flux.train!(loss,dtrn,SGD(params(m)))
    trnloss1[i] = avgloss(dtrn,m)
    tstloss1[i] = avgloss(dtst,m)
    trnerr1[i] = zeroone(dtrn,m)
    tsterr1[i] = zeroone(dtst,m)
end
```

```julia
minimum(tstloss1),minimum(tsterr1)
```

# Multi Layer Perceptron

```julia
m3 = Chain(Dense(784,64,Flux.relu), Dense(64,10))
loss3(r,s) = vecnorm(m3(r)-s)/100
```

```julia
trnloss3 = zeros(100)
tstloss3 = zeros(100)
trnerr3 = zeros(100)
tsterr3 = zeros(100)
for i = 1:100
    Flux.train!(loss3,dtrn,SGD(params(m3)))
    trnloss3[i] = avgloss(dtrn,m3)
    tstloss3[i] = avgloss(dtst,m3)
    trnerr3[i] = zeroone(dtrn,m3)
    tsterr3[i] = zeroone(dtst,m3)
end
minimum(tstloss3)
```

```julia
plot([trnloss1 tstloss1 trnloss3 tstloss3],
    labels=[:trnLin :tstLin :trnMLP :tstMLP],xlabel="Epochs",ylabel="Loss")  
# solves the underfitting problem!
```

```julia
plot([trnerr1 tsterr1 trnerr3 tsterr3],
    labels=[:trnLin :tstLin :trnMLP :tstMLP],xlabel="Epochs",ylabel="Error")  
# error improves from 8% to 2%!
```

# MLP with Dropout

```julia
m4 = Chain(Dropout(0.8),Dense(784,64,Flux.relu),Dense(64,10))
loss4(r,s) = vecnorm(m4(r)-s)/100 + 0.1*(sum(abs,m4[2].W.data)
                                        +sum(abs,m4[3].W.data)
                                        +sum(abs,m4[2].b.data)
                                        +sum(abs,m4[3].b.data))
```

```julia
trnloss4 = zeros(100)
tstloss4 = zeros(100)
trnerr4 = zeros(100)
tsterr4 = zeros(100)
for i = 1:100
    Flux.testmode!(m4, false)
    Flux.train!(loss4,dtrn,SGD(params(m4)))
    Flux.testmode!(m4)
    trnloss4[i] = avgloss(dtrn,m4)
    tstloss4[i] = avgloss(dtst,m4)
    trnerr4[i] = zeroone(dtrn,m4)
    tsterr4[i] = zeroone(dtst,m4)
end
minimum(tstloss4)
```

```julia
plot([trnloss3 tstloss3 trnloss4 tstloss4],
    labels=[:trnMLP :tstMLP :trnDropout :tstDropout],xlabel="Epochs", ylabel="Loss")  
# overfitting less, loss results improve
```

```julia
plot([trnerr3 tsterr3 trnerr4 tsterr4],
    labels=[:trnMLP :tstMLP :trnDropout :tstDropout],xlabel="Epochs", ylabel="Error")  
# this time error also improves slightly
```
