```julia
function minibatch(x,y,batchs)
    numBatch = Int((size(x)[2])/batchs)
    b = Array{Any,1}(numBatch)
    for i = 1:numBatch
        s = (i-1)*batchs + 1
        e = s + batchs -1
        b[i] = (x[:,s:e],y[:,s:e])
    end
    return b
end

function avgloss(data,model)
    loss = 0
    count = 0
    for (x,y) in data
        ypred = model(x).data
        loss += vecnorm(ypred-y,2)/100
        count += 1
    end
    return loss/count
end

function avgloss2(data,model)
    loss = 0
    count = 0
    for (x,y) in data
        loss += softloss(x,y,model)
        count += 1
    end
    return loss/count
end
```

# Polynomial Data

```julia
N = 500
L=100
x = 2*rand(N)-1;
```

```julia
a = randn(4)
function f(x)
    return 10*(x-a[1])*(x-a[2])*(x-a[3])*(x-a[4])
end
```

```julia
y = f.(x)
scatter(x,y)
```

```julia
xtrn = ones(L,N)
for i = 1:L
    xtrn[i,:] = x.^(i-1) + randn(N)
end
ytrn = y'
dtrn = minibatch(xtrn,ytrn,50);
```

```julia
xt = 2*rand(1000)-1
xtst = ones(L,1000)
for i = 1:L
    xtst[i,:] = xt.^(i-1)
end

ytst = f.(xt)'

dtst = minibatch(xtst,ytst,100);
```

```julia
m5 = Chain(Dense(L,256,Flux.tanh),Dense(256,1))
loss5(r,s) = vecnorm(m5(r)-s,2)/100
```

```julia
trnloss5 = zeros(1000)
tstloss5 = zeros(1000)

for i = 1:1000
    Flux.train!(loss5,dtrn,SGD(params(m5)))
    trnloss5[i] = avgloss(dtrn,m5)
    tstloss5[i] = avgloss(dtst,m5)
end
```

```julia
plot([trnloss5 tstloss5],labels=[:trnloss :tstloss],xlabel="Epochs",ylabel="Loss") 
```

```julia
p = 0.5
m6 = Chain(Dense(L,256,Flux.tanh),Dropout(0.5),Dense(256,1))
loss6(r,s) = vecnorm(m6(r)-s,2)/100
```

```julia
trnloss6 = zeros(1000)
tstloss6 = zeros(1000)

for i = 1:1000
    Flux.testmode!(m6, false)
    Flux.train!(loss6,dtrn,SGD(params(m6)))
    Flux.testmode!(m6)
    trnloss6[i] = avgloss(dtrn,m6)
    tstloss6[i] = avgloss(dtst,m6)
end
```

```julia
plot([trnloss5 tstloss5 trnloss6 tstloss6],labels=[:trnloss :tstloss :trnlossDrop :tstlossDrop],xlabel="Epochs",ylabel="Loss") 
```
